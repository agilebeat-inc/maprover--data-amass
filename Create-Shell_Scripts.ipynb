### A function _sh_creator(file_path, zooms)_ create shell scripts to download map tiles as .png
* **file_path** - file path where geojson file exists
* **zoom** - a list of integer between 1 ~ 19 
  (e.g.) [17] or [16, 17, 18]  

```python
import TileGenrator as TG
import math
import random
import string
```

```python
def sh_creator(file_path, zooms): 
    for zoom in zooms:
    
        file_path = file_path
        gj = TG.load_json_file(file_path)
        features = TG.extract_feature_list_from_gj(gj)
        points_list = [f["geometry"]["coordinates"] for f in features]

        points = []
        urls = []
        for p in points_list:
            x, y = TG.deg2num(p[1],p[0],zoom)
            urls.append('https://' + random.choice('abc') + '.tile.openstreetmap.org/' + str(zoom) + '/' + str(x) + '/' + str(y) + '.png')

        urls_set = set(urls)

        print("#!/bin/bash")
        counter = 0
        for u in urls_set:
            split_url = u.split('/')
            latlon = TG.num2deg(int(split_url[4]), int(split_url[5][:len(split_url[5])-4]), int(split_url[3]) )
            print('wget -O ' + str(split_url[3]) + '_' + str(split_url[4]) + '_' + str(split_url[5]) + ' ' + u \
                  + ' #Lat: ' + str(latlon[0]) + ", Lon: " + str(latlon[1]))
            if counter%50 == 0:
                print('sleep 1.1')
            counter = counter + 1
```

### A function _sh_creator()_
A function _sh_creator( )_ creates shell script from geojson file.
It requires two parameters: 
* **file_path** - a file path where geojson file exists 
* **zooms** - Zoom levels for collected tiles

```python
sh_creator('airfield_Bejing.geojson', zooms = [17,18,19])
```
